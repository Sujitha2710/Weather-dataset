```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('/content/weather_dataset_19-01-2020.csv')
temperature_data = data['country']
plt.hist(temperature_data, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('country')
plt.ylabel('wind')
plt.title('Temperature visualization')
plt.show()
```
